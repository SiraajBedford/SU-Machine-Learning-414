```python
from hmm_class import HMM
signal5 = np.array([[ 0.7, 0.5, 0.6, 2.5]])
# Define signals
signal1 = np.array([[ 1. ,  1.1,  0.9, 1.0, 0.0,  0.2,  0.1,  0.3,  3.4,  3.6,  3.5]])
signal2 = np.array([[0.8, 1.2, 0.4, 0.2, 0.15, 2.8, 3.6]])

# Collect training data together
toy_data = np.hstack([signal1, signal2])
toy_lengths = [11, 7]

# Create and fit HMM model to data
toy_hmm = HMM()
toy_hmm.fit(toy_data, toy_lengths, 3)
toy_means = [d.get_mean() for d in toy_hmm.dists]
toy_covs = [d.get_cov() for d in toy_hmm.dists]
# print ('Transition probabilities: ')
# print (toy_hmm.trans)
# print ('Means: ')
# print (toy_means)
# print ('Covariances: ')
# print (toy_covs)
```

```python
#4a
print(toy_hmm.forward(signal5))#loglikelihood
```

```python
#4b
seq, lltable = toy_hmm.viterbi(signal5)
print(seq)#most likely sequence and lltable.Returns the maximum likelihood hidden state sequence as well as the log-likelihood of that sequence.
```

```python
#4c
print(toy_hmm.score(signal5, np.array([0, 1, 1, 2]) ))#l Returns the log-likelihood of the observation-states correspondence.
print(lltable)
```
